```python
 def sentimentalAnalysis(data,columnname):
        
        #downloading vader_lexicon for the process
        import matplotlib.pyplot as plt
        import seaborn as sns
        "Importing Necessary Packeage"
        from nltk.sentiment.vader import SentimentIntensityAnalyzer
        sid = SentimentIntensityAnalyzer()
        
        "Deleting null pr empty value"
        data.dropna(inplace=True)
        
        "Checking for a comment"
        #sid.polarity_scores(data[columnname][93])
        
        "Creating respective columns"
        
        data['scores'] = data[columnname].apply(lambda commentText: sid.polarity_scores(commentText))
        data['compound']  = data['scores'].apply(lambda score_dict: score_dict['compound'])
        data['Negtive']  = data['scores'].apply(lambda score_dict: score_dict['neg'])
        data['Postive']  = data['scores'].apply(lambda score_dict: score_dict['pos'])
        data['Neutral']  = data['scores'].apply(lambda score_dict: score_dict['neu'])
        
        "Creating final pos or neg using compound score"
        data['comp_score'] = data['compound'].apply(lambda c: 'pos' if c >=0 else 'neg')
        plt.clf()
        #comp=sns.countplot(x = 'comp_score', hue = 'Name', data = data, palette = 'magma')
        #comp.figure.savefig("date_charts.png")
        "Checking how many pos and neg"
        posneg=pd.DataFrame(data['comp_score'].value_counts())
        return posneg,data
```

```python
pos,data_Senti=sentimentalAnalysis(dataset,columnname='Chat')
```

# Identifying missed voice and video calls 

```python

# Placeholder function to identify media files
def is_media(chat):
    if '<Media omitted>' in chat:
        return 1
    else:
        return 0

# Placeholder function to identify deleted messages
def is_deleted_message(chat):
    if 'This message was deleted' in chat:
        return 1
    else:
        return 0

# Function to identify missed voice calls
def is_missed_voice_call(chat):
    if 'Missed a voice call' in chat:
        return 1
    else:
        return 0

# Function to identify missed video calls
def is_missed_video_call(chat):
    if 'Missed a video call' in chat:
        return 1
    else:
        return 0

# Apply functions to calculate metrics
df['Media_Count'] = df['Chat'].apply(is_media)
df['Deleted_Message_Count'] = df['Chat'].apply(is_deleted_message)
df['Missed_Voice_Call_Count'] = df['Chat'].apply(is_missed_voice_call)
df['Missed_Video_Call_Count'] = df['Chat'].apply(is_missed_video_call)

# Calculate total counts
media_count = df['Media_Count'].sum()
deleted_message_count = df['Deleted_Message_Count'].sum()
missed_voice_call_count = df['Missed_Voice_Call_Count'].sum()
missed_video_call_count = df['Missed_Video_Call_Count'].sum()

# Print results
print("Media count:", media_count)
print("Deleted message count:", deleted_message_count)
print("Missed voice call count:", missed_voice_call_count)
print("Missed video call count:", missed_video_call_count)
```

#  Detect flirt encounters

```python
import re
# Function to detect flirt encounters
def detect_flirt_encounters(chat):
    flirt_patterns = [
        r'(?i)\b(?:cute|hot|sexy|beautiful|gorgeous)\b',
        r'(?i)\b(?:love|miss|thinking about|dreaming about)\b',
        r'(?i)\b(?:date|coffee|dinner|movie)\b'
    ]
    for pattern in flirt_patterns:
        if re.search(pattern, chat):
            return 1
    return 0

# Apply function to detect flirt encounters
df['Flirt_Encounter'] = df['Chat'].apply(detect_flirt_encounters)

# Calculate total flirt encounters and total messages for each person
flirt_data = df.groupby('Name').agg({'Flirt_Encounter': 'sum', 'Chat': 'count'}).reset_index()
flirt_data.columns = ['Name', 'Flirt_Count', 'Total_Messages']

# Calculate flirt percentage for each person
flirt_data['Flirt_Percentage'] = (flirt_data['Flirt_Count'] / flirt_data['Total_Messages']) * 100

# Print results
print(flirt_data)
```
